```python
liss = []
links = []
for i in range(len(div1)):
    #print(div1[i].text.split('\n')[1:-1] , div1[1].a['data-vdplink'])
    liss.append(div1[i].text.split('\n')[1:-1])
    links.append(div1[1].a['data-vdplink'])
```

```python
liss1 = []
div1 = soup.findAll('div' , {"class" : "lst-dtl clearfix"})
for i in range(len(div1)):
    #print(div1[i].text.split('\n')[1:4])
    liss1.append(div1[i].text.split('\n')[1:4])
```

```python
t = []
for i in range(len(liss)):
    t.append(liss[i]+liss1[i]+ [links[i]])
```
